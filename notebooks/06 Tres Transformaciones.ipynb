#### Contenido adicional calculo del angulo entre vectores

Dada la formula

A * B = _norma(A)_ \* _norma(B)_ \* $\cos$($\alpha$)

Despejando 

$\cos$($\alpha$) = $\frac{A * B}{norma(A) *norma(B) }$ 

```
B = np.array([[1, 0],
       [0, 1]])
print(B)
```

```
print(U)
```

```
cosAngRadian = U[0,:].dot(B[0,:]) / np.linalg.norm(U[0,:]) * np.linalg.norm(B[0,:])
print(cosAngRadian)
```

```
# Obtenemos el angulo
angRadian = np.arccos(cosAngRadian)
print(angRadian)
```

```
# Conversion del angulo a grados
np.rad2deg(angRadian)
```
