Un autovector es aquel que cuando le aplico una matriz me devuelve el vector con la misma direcci√≥n pero puede tener una amplitud distinta. Puede estar multiplicado por el autovalor.

```
# Este valor no es unico, podemos tener otro
s = np.array([[-1],[2]])
print(s)
```

```
t = x.dot(s)
print(t)
```

```
graficarVectores([t.flatten(), s.flatten()],
                  cols=[orange_light, blue_light])

plt.xlim(-3, 3)
plt.ylim(-3, 3)
```

La matriz cambio de sentido, mas no de direccion, el vector ahora apunta en el sentido opuesto

Con esto vemos que una matriz tiene dos autovectores con autovalores asociados
